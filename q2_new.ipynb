```

msk = np.random.rand(len(train_data2)) < 0.8
train = train_data2[msk].reset_index()
validate = train_data2[~msk].reset_index()

t_label = train.iloc[:,1].to_list()#column1 of train data
#print(t_label)
v_label = validate.iloc[:,1].to_list()#column1 of validate data
#print(v_label)

train_data_t = train.iloc[:,2:]
print(train_data_t.shape)
validation_set_v = validate.iloc[:,2:]
print(validation_set_v.shape)
```

```
#encoding labels

t_label1=[]
for r in t_label:
    if r=='p':
        t_label1.append(1)
    else:
        t_label1.append(0)
#print(t_label1)
v_label1=[]
for r in v_label:
    if r=='p':
        v_label1.append(1)
    else:
        v_label1.append(0)  
#print(v_label1)
```

```

def dist_manhattan(x1,x2):
    return np.sum(np.absolute(x1-x2))

def count2(pred,k):
    x = pred[1]
    return (np.bincount(x).argmax())

def KNN(train_data,query_point,label,k=7):
    val = []
#     print(label)
    for ix in range(train_data.shape[0]):
        v = dist_manhattan(query_point,train_data.loc[ix, :])
#         print(label[ix])
        v1 =[v,label[ix]]
        #print(v)
        val.append(v1)
        
        
    updated_val = sorted(val,key=lambda x : x[0]) 
    #print(type(updated_val))
    pred_ar = (updated_val[:k]) 
    #print(pred_ar)#list
    return count2(pred_ar,k)
```

```
predicted_label = []
print(validation_set_v.shape)
for i in range(validation_set_v.shape[0]):
    query = validation_set_v.loc[i,:] 

    res = KNN(train_data_t,query,t_label1)#sample =1 for euclidean and for any other value its any value
    predicted_label.append(res)
    print("appending")
print(predicted_label)
```
